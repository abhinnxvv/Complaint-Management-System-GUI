```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
desk_data=pd.read_csv("desktop.csv")
prin_data=pd.read_csv("printer.csv")
prin_data2=pd.read_csv("printer2.csv")
```

```python
val=int(input('''ENTER 
         1: Desktop issued by IT Department
         2: Printer issued by IT Department
         3: Printer Details of UPMRC\n'''))
if val == 1:
    desk_data.fillna("N/A")
    desk_data.sample(10)
    TFT=input("Enter the TFT value ")
    DATA=desk_data.loc[desk_data["TFT"]==TFT,:] 
    lst1=DATA.values
    print("CPU:       ",lst1[0][3])
    print("UPS:       ",lst1[0][4])
    print("Issued to: ",lst1[0][5])
    print("Department:",lst1[0][6])
    print("Query     :",lst1[0][13])
    index=int(lst1[0][0])-1
    query=input("Want to enter a query?y/Y")
    try:
        if(query=="y" or query=="Y"):
            query_input=input("Enter the query")
            desk_data.at[index, 'QUERY'] = query_input
            desk_data.to_csv("desktop.csv", index=False)
    except:
        pass
if val==2:
    prin_data.fillna("N/A")
    prin_data.sample(10)
    Serial=input("Enter the Printer serial number ")
    DATA=prin_data[prin_data["Product No/Serial No"]==Serial] 
    lst1=DATA.values

    print("Issued to :",lst1[0][4])
    print("Department:",lst1[0][5])
    print("Location  :",lst1[0][9])
    print("Query     :",lst1[0][14])

    query=input("Want to enter a query?y/Y")
    try:
        if(query=="y" or query=="Y"):
            query_input=input("Enter the query ")
            index=int(lst1[0][0])-1
            prin_data.at[index, 'QUERY'] = query_input
            prin_data.to_csv("printer.csv", index=False)
    except:
        pass
if val==3:
    prin_data2.fillna("N/A")
    prin_data2.sample(10)
    Model=input("Enter the Printer Model ")
    DATA=prin_data2[prin_data2["Model"]==Model] 
    lst1=DATA.values
    for i in range(len(lst1)):
        print("Issued to :",lst1[i][3])
        print("Department:",lst1[i][5])
        print("Query     :",lst1[i][6])

        print("\n\n")
    query=input("Want to enter a query?y/Y")
    try:
        if(query=="y" or query=="Y"):
            query_input=input("Enter the query ")
            for i in range(len(lst1)):
                index=int(lst1[i][0])-1
                prin_data2.at[index, 'QUERY'] = query_input
            prin_data2.to_csv("printer2.csv", index=False)
    except:
        pass
```

```python
import tkinter as tk
from tkinter import messagebox

def get_desktop_data():
    desk_data.fillna("N/A")
    desk_data.sample(10)
    TFT = input_entry.get()
    DATA = desk_data.loc[desk_data["TFT"] == TFT, :]
    lst1 = DATA.values

    if len(lst1) > 0:
        result_text = "CPU:       {}\n".format(lst1[0][3])
        result_text += "UPS:       {}\n".format(lst1[0][4])
        result_text += "Issued to: {}\n".format(lst1[0][5])
        result_text += "Department: {}\n".format(lst1[0][6])
        result_text += "Query:     {}".format(lst1[0][13])

        index = int(lst1[0][0]) - 1
        query = messagebox.askyesno("Enter Query", "Want to enter a query?")
        try:
            if query:
                query_input = query_entry.get()
                desk_data.at[index, 'QUERY'] = query_input
                desk_data.to_csv("desktop.csv", index=False)

            messagebox.showinfo("Result", result_text)
        except:
            messagebox.showerror("Error", "No data found for the given TFT value.")
        

def get_printer_data():
    prin_data.fillna("N/A")
    prin_data.sample(10)
    Serial = input_entry.get()
    DATA = prin_data[prin_data["Product No/Serial No"] == Serial]
    lst1 = DATA.values

    if len(lst1) > 0:
        result_text = "Issued to : {}\n".format(lst1[0][4])
        result_text += "Department: {}\n".format(lst1[0][5])
        result_text += "Location  : {}\n".format(lst1[0][9])
        result_text += "Query     : {}".format(lst1[0][14])

        query = messagebox.askyesno("Enter Query", "Want to enter a query?")
        try:
            if query:
                query_input = query_entry.get()
                index = int(lst1[0][0]) - 1
                prin_data.at[index, 'QUERY'] = query_input
                prin_data.to_csv("printer.csv", index=False)

            messagebox.showinfo("Result", result_text)
        except:
            messagebox.showerror("Error", "No data found for the given Printer serial number.")

def get_upmrc_printer_data():
    prin_data2.fillna("N/A")
    prin_data2.sample(10)
    Model = input_entry.get()
    DATA = prin_data2[prin_data2["Model"] == Model]
    lst1 = DATA.values

    if len(lst1) > 0:
        result_text = ""
        for i in range(len(lst1)):
            result_text += "Issued to : {}\n".format(lst1[i][3])
            result_text += "Department: {}\n".format(lst1[i][5])
            result_text += "Query     : {}\n\n".format(lst1[i][6])

        query = messagebox.askyesno("Enter Query", "Want to enter a query?")
        try:
            if query:
                query_input = query_entry.get()
                for i in range(len(lst1)):
                    index = int(lst1[i][0]) - 1
                    prin_data2.at[index, 'QUERY'] = query_input
                prin_data2.to_csv("printer2.csv", index=False)

            messagebox.showinfo("Result", result_text)
        except:
            messagebox.showerror("Error", "No data found for the given Printer Model.")

def select_option():
    val = option_var.get()
    if val == 1:
        get_desktop_data()
    elif val == 2:
        get_printer_data()
    elif val == 3:
        get_upmrc_printer_data()

# Create the Tkinter window
window = tk.Tk()
window.title("Data Retrieval")
window.geometry("400x300")
input_label = tk.Label(window, text="\nENTER \n1: Desktop issued by IT Department\n2: Printer issued by IT Department\n3: Printer Details of UPMRC\n")
input_label.pack()

# Create the option dropdown menu
option_var = tk.IntVar()
option_var.set(1)

option_label = tk.Label(window, text="Select an option:")
option_label.pack()

option_menu = tk.OptionMenu(window, option_var, 1, 2, 3)
option_menu.pack()

# Create the input entry field
input_label = tk.Label(window, text="Enter:")
input_label.pack()

input_entry = tk.Entry(window)
input_entry.pack()

search_button = tk.Button(window, text="QUERY", command=select_option)
search_button.pack()


# Create the query entry field
input_label = tk.Label(window, text="Enter a Query:")
input_label.pack()

query_entry = tk.Entry(window)
query_entry.pack()

# Create the button
search_button = tk.Button(window, text="DONE", command=select_option)
search_button.pack()

search_button = tk.Button(window, text="SUBMIT", command=select_option)
search_button.pack()

# Run the Tkinter main loop
window.mainloop()
```
